**Q1: Explain the roles of the Driver, Cluster Manager, and Executor in a Spark application?**

Ans:In Apache Spark, the Driver acts like the brain of the application. It controls the overall execution, creates the SparkSession, breaks the code into smaller tasks, and decides when and where those tasks should run.

The Cluster Manager handles resource management across the cluster, such as CPU and memory allocation. Its job is to make sure Spark applications get the resources they need to run efficiently. Common examples include Apache Hadoop YARN, Apache Mesos, and Kubernetes.

The Executors are the worker processes that actually perform the tasks assigned by the Driver. They handle computations and can also store intermediate data in memory or on disk, which helps speed up data processing.

**Q2: How does Spark’s Lazy Evaluation strategy improve performance when chain-processing large datasets?**

Ans:In Apache Spark, Lazy Evaluation means transformations such as filter() and select() are not executed right away. Instead, Spark simply keeps track of these operations and waits until an action like show() or collect() is called before processing the data.

This approach improves performance because Spark gets time to optimize the execution plan. It can merge multiple transformations, reduce unnecessary data movement, and use resources more efficiently. As a result, processing large datasets becomes faster and more efficient.

**Q4: What is the difference between CSV and Parquet in terms of storage (row-based vs. columnar) and why does it matter for performance?**

Ans: CSV is a row-based file format, meaning data is stored one row after another. It is simple and easy to read, but it can be slower when handling large amounts of data.

Parquet is a column-based file format, where data is stored column by column instead of row by row. It supports compression and is optimized for analytical workloads, making queries faster.

In Apache Spark, Parquet usually performs better because Spark can read only the columns needed for a query rather than scanning the entire dataset. This reduces I/O operations and improves processing speed.

**Q7: How does Spark use the Lineage Graph (DAG) to provide fault tolerance if a worker node fails?**

Ans: In Apache Spark, Spark maintains a DAG (Directed Acyclic Graph) that keeps track of all the transformations applied to a dataset.

If a worker node fails and some data gets lost, Spark uses this DAG to rebuild only the missing partitions instead of processing the entire dataset again.

This makes Spark highly fault-tolerant and ensures efficient recovery with minimal performance impact.

**Q15: When exploring a dataset, why is it safer to use .show(5) instead of .collect() on a multi-terabyte dataset?**

Ans: .collect() brings the entire dataset into the Driver’s memory.

For very large datasets, this can cause problems such as:

Memory overflow
Slow execution
Application crash

On the other hand, .show(5) displays only the first 5 rows of the dataset. This makes it much safer and more efficient for exploring data.Therefore, .show() is generally preferred over .collect() when working with large datasets because it uses less memory and helps keep the system stable.
