 6.Create a derived column (total_amount = price * quantity).

```python
import pandas as pd
csv=pd.read_csv("/content/Combined_dataset.csv")
csv
```

```python
csv["total_amount"]=csv["final_price"]* csv["ratings_count"]
```

```python
csv.drop([
    "currency",
    "images",
    "amount_of_stars",
    "what_customers_said",
    "seller_name",
    "sizes",
    "videos",
    "seller_information",
    "variations",
    "best_offer",
    "more_offers"
],axis=1,inplace=True)
```

```python
csv
```

```python
csv.rename(columns={"ratings_count" : "quantity"},inplace=True)
```

```python
csv.columns
```

```python
csv["final_price"] = (
    csv["final_price"]
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.replace('"', "", regex=False)
    .astype(float)
)
```

```python
csv["final_price"]=csv["final_price"].astype(int)
```

```python
csv["quantity"]=csv["quantity"].astype(int)
```

```python
csv
```

```python
csv["total_amount"]=csv["final_price"]* csv["quantity"]
```

```python
csv.total_amount
```

```python
csv
```

```python
csv.drop ([
    "delivery_options",
    "product_details",
    "breadcrumbs",
    "product_specifications"
],axis=1,inplace=True)
```

```python
csv
```

```python
csv.to_csv("Cleaned_Data.csv",index=False)
```
