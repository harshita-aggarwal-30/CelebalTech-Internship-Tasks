Clean Customers Dataset

```python
def clean_customers(df):

    # Remove extra spaces
    df["customer_name"] = df["customer_name"].str.strip()

    # Convert email to lowercase
    df["email"] = df["email"].str.lower()

    # Fill missing email
    df["email"] = df["email"].fillna("unknown@gmail.com")

    return df

customers = clean_customers(customers)
```

```python
customers.head(10)
```

Validate Emails

```python
def validate_emails(df):

    invalid_email = df[~df["email"].str.contains("@", na=False)]

    return invalid_email

invalid_email_df = validate_emails(customers)

print("Number of Invalid Emails :", len(invalid_email_df))

invalid_email_df
```

Fix Invalid Emails

```python
customers["email"] = customers["email"].apply(
    lambda x: "unknown@gmail.com" if "@" not in str(x) else x
)

customers["email"].str.contains("@").value_counts()
```

Generate Summary Report

```python
summary_query = """
SELECT
COUNT(DISTINCT o.order_id) AS total_orders,
ROUND(SUM(oi.quantity * oi.unit_price * (1 - oi.discount_percent / 100)),2) AS total_revenue,
COUNT(DISTINCT o.customer_id) AS unique_customers
FROM orders o
JOIN order_items oi
ON o.order_id = oi.order_id
WHERE DATE(o.order_date)
BETWEEN %s AND %s;
"""

cursor.execute(summary_query, (start_date, end_date))

summary = cursor.fetchone()


print("Report Type       :", report_type)
print("Date Range        :", start_date, "to", end_date)


print("Total Orders      :", summary[0])
print("Total Revenue     : ₹", summary[1])
print("Unique Customers  :", summary[2])
```

Top 3 Products

```python
top_query = """
SELECT
p.product_name,
SUM(oi.quantity) AS total_quantity
FROM products p
JOIN order_items oi
ON p.product_id = oi.product_id
JOIN orders o
ON oi.order_id = o.order_id
WHERE DATE(o.order_date)
BETWEEN %s AND %s
GROUP BY p.product_name
ORDER BY total_quantity DESC
LIMIT 3;
"""

cursor.execute(top_query, (start_date, end_date))

top_products = cursor.fetchall()

for i, product in enumerate(top_products, start=1):
    print(f"{i}. {product[0]} - {product[1]} Units")
```

Test 1 : Invalid Order ID

1. What happens when order_items has an order_id not in orders? 

```python
def test_invalid_order_id():

    query = """
    SELECT *
    FROM order_items
    WHERE order_id NOT IN (
        SELECT order_id
        FROM orders
    );
    """

    cursor.execute(query)

    result = cursor.fetchall()

    print("\nTest 1 : Invalid Order ID")
    print("Records Found :", len(result))

test_invalid_order_id()
```

2. What happens when discount_percent > 100? 

```python
def test_discount():

    query = """
    SELECT *
    FROM order_items
    WHERE discount_percent > 100;
    """

    cursor.execute(query)

    result = cursor.fetchall()

    print("\nTest 2 : Discount Greater Than 100")
    print("Records Found :", len(result))

test_discount()
```

3. What happens when quantity is 0? 

```python
def test_zero_quantity():

    query = """
    SELECT *
    FROM order_items
    WHERE quantity = 0;
    """

    cursor.execute(query)

    result = cursor.fetchall()

    print("\nTest 3 : Quantity Equal to Zero")
    print("Records Found :", len(result))

test_zero_quantity()
```

4. What happens when order_date is in the future? 

```python
def test_future_date():

    query = """
    SELECT *
    FROM orders
    WHERE order_date > CURDATE();
    """

    cursor.execute(query)

    result = cursor.fetchall()

    print("\nTest 4 : Future Order Date")
    print("Records Found :", len(result))

test_future_date()
```
