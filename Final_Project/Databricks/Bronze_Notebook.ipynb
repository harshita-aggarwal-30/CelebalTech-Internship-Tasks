#  Phase 1 - Bronze Layer(Load the raw data)

**Import Required Libraries**

```python
from pyspark.sql import SparkSession
```

**Define File Paths**

```python
customers_path = "/Volumes/workspace/default/servicetrack_data/customers.csv"

devices_path = "/Volumes/workspace/default/servicetrack_data/devices.csv"

jobs_path = "/Volumes/workspace/default/servicetrack_data/service_jobs.csv"
```

**Load the CSV Files**

```python
df_customers = spark.read.csv(
    customers_path,
    header=True,
    inferSchema=True
)

df_devices = spark.read.csv(
    devices_path,
    header=True,
    inferSchema=True
)

df_jobs = spark.read.csv(
    jobs_path,
    header=True,
    inferSchema=True
)
```

**Display the Data**

```python
display(df_customers)
```

**Check the Schema**

```python
df_customers.printSchema()

df_devices.printSchema()

df_jobs.printSchema()
```

**Count Records**

```python
print("Customers :", df_customers.count())

print("Devices :", df_devices.count())

print("Jobs :", df_jobs.count())
```

**Check for Missing Values**

```python
display(df_customers.describe())
```

```python
display(df_devices.describe())
```

```python
display(df_jobs.describe())
```

**Save Data as Bronze Tables**

```python
df_customers.write \
.format("delta") \
.mode("overwrite") \
.saveAsTable("workspace.default.bronze_customers")
```

```python
df_devices.write \
.format("delta") \
.mode("overwrite") \
.saveAsTable("workspace.default.bronze_devices")
```

```python
df_jobs.write \
.format("delta") \
.mode("overwrite") \
.saveAsTable("workspace.default.bronze_service_jobs")
```

**Verify the Bronze Tables**

```python
bronze_customers = spark.table("workspace.default.bronze_customers")

display(bronze_customers)
```

**Brief insights about Why we create bronze layer and what actions we performed under this :-**

Q1. What is the purpose of the Bronze Layer?
Ans: Bronze layer act as a backup layer . we load the raw data as it is in bronze layer without any modifications. It ensures that no data or information is lost before data cleaning and transformations.

Q2. Why did we use Delta tables?
Ans: Delta tables are used to store data in an efficient way.They support ACID transactions, which help keep the data accurate and consistent.In addition, they offer better performance for reading and writing data.They are commonly used in Databricks to build Medallion Architecture pipelines.

Q3. Why did we use inferSchema=True?
Ans: allows Apache Spark to automatically identify the correct data type for each column.This reduces manual effort, minimizes data type errors, and makes data processing and analysis more accurate and efficient.
